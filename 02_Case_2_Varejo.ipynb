```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
#Apagando linhas com nulos
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

Os dois abaixo tem a mesma função.

Um atribuo antes que a coluna vendas['estado]' = ...
No outro, uso inplace = True

```python
vendas['estado'] = vendas['estado'].fillna('MS')
vendas.info()
```

```python
vendas['estado'].fillna('MS', inplace = True)
vendas.info()
```

```python
#Com a função mean() tirei a média da coluna preço do meu dataset vendas e atribuí a uma variável, media_preco

media_preco = vendas['Preço'].mean()


#Utilizando fillna() para preencher os vazios com minha nova variável, conforme acima, media_preco

vendas['Preço'].fillna(media_preco, inplace = True)
vendas.info()
```

```python
media_preco
```

```python
mediana_preco = vendas['Preço'].median()
mediana_preco
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
#Verificando se preço é maior do que preço com frete

preco_errado = vendas.query('Preço > Preço_com_frete')
vendas_correto = vendas.query('Preço < Preço_com_frete')

vendas_correto.describe()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
#Departamentos mais vendidos

vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
#Média de preços com frete por departamento

round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg('mean').sort_values(ascending = False).reset_index(),2)
```

```python
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
vendas_correto['mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending = False).reset_index()
```

# Base de Clientes

```python
cliente = pd.read_excel('cliente_varejo.xlsx')
```

```python
cliente[['idade','renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how= "left", on="cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color = 'green') #Eixo X, depois Y
plt.ylabel = ('Média de Idade')
plt.title('Idade Média por Bandeira', loc = 'center', fontsize = 18, fontweight = 'bold', color = 'black');
```

```python
#EXTRA
#Exibir a renda média por aplicativo e plotar um gráfico de barras

agg_media_renda_apps = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_media_renda_apps
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_media_renda_apps['idcanalvenda'], agg_media_renda_apps['renda'], color = 'green')
plt.ylabel = ('Média de Renda')
plt.title('Renda Média por App', loc = 'center', fontsize = 18, fontweight = 'bold', color = 'black')
```

```python
vendas_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize=(15,4))

plt.xlabel = ("Data da Venda")
plt.ylabel = ("Quantidade de Vendas")
x = vendas_por_data["Data"]
y = vendas_por_data['idcompra']

plt.plot(x,y, color = 'green', linewidth = 2)
plt.show()
```

```python
media_precofrete_dep = round(vendas_cliente.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending = False).reset_index(),2)
media_precofrete_dep
```

```python
fig = plt.figure(figsize = (20,4))

plt.xlabel = ("Departamento")
plt.ylabel = ("Preço com Frete")
x = media_precofrete_dep['Nome_Departamento']
y = media_precofrete_dep['Preço_com_frete']

plt.bar(x,y, color = 'blue')
plt.show()
```

```python
fig = plt.figure(figsize = (30,20))

plt.bar(media_precofrete_dep["Nome_Departamento"], media_precofrete_dep["Preço_com_frete"])
plt.xlabel = ("Departamento")
plt.ylabel = ("Preço com Frete")
```
